# Pipelines
I created two pipelines one with a ridge regressor and one with a xgboost regressor. 
Below you see the visualisations of both of the. 

Both share the same feature extraction pipeline. The feature tranformation pipeline is different though. 

More on the feautre extraction pipeline can be found in the EDA notebook.

## Ridge
Ridge pipeline is trained using CV to find the best alpha. 

The following feature/target tranformations are applied: 
* All boolean features are transformed to 0/1
* All categorical features are transformed to dummy binary features
* Skwewed variables and the target Sales are unsked using the PowerTransfomer
* The time features are scales using the MinMax Scaler



```python
ridge_pipeline = ridge.make_pipeline(prepared_stores)
ridge_pipeline
```
